```python
#Длина сессий неплатящего пользователя
wow_new_df_np=df_1_final[[ 'Day', 'Videolength']]
wow_new_df_np=wow_new_df_np.drop_duplicates(ignore_index=False)
#wow_new_df_np['Videolength']=pd.to_datetime(wow_new_df_np['Videolength'])
#wow_new_df_np['Videolength']=pd.to_timedelta(wow_new_df_np['Videolength']).astype(float)
wow_new_df_np['Videolength']=wow_new_df_np['Videolength'].astype(str)
wow_new_df_np['Videolength']=wow_new_df_np['Videolength'].str.replace(":", " ")
wow_new_df_np['Videolength']=wow_new_df_np['Videolength'].str.replace(" ", "")
#wow_new_df_np['Videolength']=pd.to_numeric(wow_new_df_np['Videolength'])
wow_new_df_np['Videolength']=wow_new_df_np['Videolength'].astype(float)
wow_new_df_np['Videolength']=wow_new_df_np['Videolength']/10000
wow_new_df_np=wow_new_df_np.groupby(['Day']).agg({'Videolength':['sum']})
wow_new_df_np
```
